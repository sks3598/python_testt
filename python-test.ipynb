```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from nsepy import get_history
from datetime import date
data = get_history(symbol="SBIN", start=date(2015,1,1), end=date(2016,1,31)) 
```

```python
data.head()
```

```python
data.describe()
```

```python
data_close = pd.DataFrame(data.Close)
```

```python
data_close["m_avg_10"] = data_close.Close.rolling(10).mean()
data_close["m_avg_75"] = data_close.Close.rolling(75,center=True).mean()
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(data_close['Close'],label="close")
plt.plot(data_close['m_avg_10'],label="Avg 10 day")
plt.plot(data_close['m_avg_75'],label="Avg 75 day")
plt.legend(loc=2)
```
